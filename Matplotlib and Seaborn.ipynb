# Data Visualization Using Python

Python offers several plotting libraries, namely Matplotlib, Seaborn and many other such data visualization packages with different features for creating informative, customized, and appealing plots to present data in the most simple and effective way.

# Matplotlib and Seaborn

Matplotlip and Seaborn are python libraries that are used for data visualization. They have inbuilt modules for plotting different graphs. While Matplotlib is used to embed graphs into applications, Seaborn is primaly used for statistical graphs.

But When should we use either of the two? Let's Understand this with the help of a comparative analysis. Below are the comparison between Python's two well-known visualization packages Matplotlib and Seaborn.

# Matplotlib

-It is used for basic graph plotting like line charts, bar graphic, etc.

-It mainly works with datasets and arrays.

-Matplotlib acts productively with data arrays and frames. It regards the aces and figures as objects.

-Matplotlib is more customizable and pairs well with Pandas and Numpy for Exploratory Data Analys.

# Seaborn

-It is mainly used for statistics visualization and can perform complex visualizations with fewer commands.

-It works with entire datasets.

-Seaborn is considerably more organized and functional than Matplotlib and treats the entire dataset as a solitary unit.

-Seaborn has more inbuilt themes and is mainly used for statistical analysis.

# Line Chart
A Line chart is a graph that represents information as a series of data points connected by a straight line. In line charts, each data point or marker is plotted and connected with a line curve.

Let's consider the apple yield (tons per hectare) in Kanto. Let's plot a line graph using this data and see how the yield of apples changes over time. We are going to both plot by using Matplotlib and Seaborn as follows :

# Using Matplotlib
We are using random data points to represent the yield of apples.

```python
yield_apples = [0.89, 0.91, 0.92, 0.93, 0.94, 0.95] #ton per hectare
```

after we define the data points of apple yields, we are going to plot its Line Chart using Matplotlib

```python
plt.plot(yield_apples)
```

To better understand the graph and its purpose, we can add the x-axis values too.

```python
years = [2017, 2018, 2019, 2020, 2021, 2022]
yield_apples = [0.89, 0.91, 0.92, 0.93, 0.94, 0.95]
```

Since we have the 'years' variable as the x-axis, we need to type it before the yield_apples variable below

```python
plt.plot(years, yield_apples) #plt.plot(x, y)
```

```python
plt.plot(years, yield_apples)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
```

To plot multiple datasets on the same graph, just use the plt,plot function once for each dataset. Let's use this to compare of yields of apples vs. oranges on the sampe graph

```python
apples = [0.89, 0.91, 0.93, 0.97, 0.986, 0.943]
tangerines = [0.79, 0.61, 0.93, 0.97, 0.962, 0.931]
```

```python
plt.plot(years, apples)
plt.plot(years, tangerines)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
```

We can add a legend which tells us what each line in our graph means. To understand what we are plotting, we can add a title to our graph

```python
plt.plot(years, apples)
plt.plot(years, tangerines)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Tangerines'])
```

To show each data point on our graph, we can highlight them with markers using the marker argument. different marker shapes like a circle, cross, square, diamond, etc. are provided by Matplotlib.

```python
plt.plot(years, apples, marker='o')
plt.plot(years, tangerines, marker='x')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Tangerines'])
```

You can use the plt.figure function to change the size of the figgure

```python
plt.figure(figsize=(12,6))

plt.plot(years, tangerines, marker = 'o')
plt.title("Yield of Tangerines (tons per hectare)")
```

## Bar Graph
When you have categorical data, you can present it with a bar graph. A bar graph plots data with the help of bars, which represent value on the y-axis and category on the x-axis. Bar graphs use bars with varying heights to show the data which belongs to a specific category.

### Using Matplotlib

```python
plt.bar(years, tangerines) #plt.bar(x, y)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
```

## Histograms
A Histogram is a bar representation of data that varies over a range. It plots the height of the data belonging to a range along the y-axis and the range along the x-axis. Histograms are used to plot data over a range of values. They use a bar representation to show the data belonging to each range. Let's again use the 'iris' data which contains information about flowers to plot histograms.

The 'iris' datasest provided by Seaborn which is completely numerical.

```python
iris_df = sns.load_dataset("iris")
iris_df.head(3)
```

### Using Matplotlib
We are going to plot a histogram from the 'Sepal Width' column.

```python
iris_df.sepal_width
```

```python
plt.title("Distribution of Sepal Width")
plt.hist(iris_df.sepal_width);
```

We can control the number or size of bins too.

```python
#specifying the number of bins
plt.hist(iris_df.sepal_width, bins=5);
```

# 20-10-2023

```python
import numpy as np

#specifying the number of bins
plt.hist(iris_df.sepal_width, bins=np.arange(2,5,0.25));
```

we can create bins of unequal size too

```python
#bins of unequal sizes
plt.hist(iris_df.sepal_width, bins=[1,3,4,4.5])
```

Similiar to line charts, we can draw multiple histograms in a single chart. We can reduce each histogram's opacity so that one histogram's bars don't hide the others. Let's draw separate histograms for each species of flowers.

```python
setosa_df = iris_df[iris_df.species == 'setosa']
versicolor_df = iris_df[iris_df.species == 'versicolor']
virginica_df = iris_df[iris_df.species == 'virginica']
```

```python
plt.hist(setosa_df.sepal_width, alpha = 0.4)
plt.hist(versicolor_df.sepal_width, alpha = 0.4)
```

Multiple Hisstograms can be stacked on top of one another by setting the stacked parameter to True.

```python
plt.title('Distribution of Sepal Width')

plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width],
        stacked=True)
plt.legend(['Setosa', 'Versicolor', 'Virginica'])
```
